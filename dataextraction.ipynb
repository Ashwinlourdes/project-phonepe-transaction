```python
import os
#Aggregate insurance path
Aggr_insur_path="D:\\phonepe project\\pulse\\data\\aggregated\\insurance\\country\\india\\state\\"
Agg_insur_list=os.listdir(Aggr_insur_path)
```

```python
import pandas as pd
import json
import os
#Aggregated insurance data
Agg_ins_data={'States':[], 'Years':[],'Quarter':[],'Transaction_type':[], 'Insurance_count':[], 'Insurance_amount':[]}

for i in Agg_insur_list:
    p_i=Aggr_insur_path+i+"\\"
    Agg_yr=os.listdir(p_i)
    for j in Agg_yr:
        p_j=p_i+j+"\\"
        Agg_yr_list=os.listdir(p_j)
        for k in Agg_yr_list:
            p_k=p_j+k
            Data=open(p_k,'r')
            D=json.load(Data)
            for z in D['data']['transactionData']:
              Name=z['name']
              count=z['paymentInstruments'][0]['count']
              amount=z['paymentInstruments'][0]['amount']
              Agg_ins_data['Transaction_type'].append(Name)
              Agg_ins_data['Insurance_count'].append(count)
              Agg_ins_data['Insurance_amount'].append(amount)
              Agg_ins_data['States'].append(i)
              Agg_ins_data['Years'].append(j)
              Agg_ins_data['Quarter'].append(int(k.strip('.json')))
#Succesfully created a dataframe
Agg_Insurance_data=pd.DataFrame(Agg_ins_data)
```

```python

district_top_ins={'States':[] ,'Years':[],'Quarter':[],'District':[],'Transaction_type':[],'Transaction_count':[],'Transaction_amount':[]}

pincode_top_ins = {'States':[] ,'Years':[],'Quarter':[],'Pincodes':[],'Transaction_type':[],'Transaction_count':[],'Transaction_amount':[]}

for i in top_insur_list:
    p_i=top_insur_path+i+"/"
    Agg_yr=os.listdir(p_i)
    for j in Agg_yr:
        p_j=p_i+j+"/"
        Agg_yr_list=os.listdir(p_j)
        for k in Agg_yr_list:
            p_k =p_j+k
            with open(p_k, 'r') as Data:
                D = json.load(Data)
            for x,y in D['data'].items():
                if x == 'districts' and y :
                    for m in y:
                        district_top_ins['States'].append(i)
                        district_top_ins['Years'].append(j)
                        district_top_ins['Quarter'].append(int(k.strip('.json')))
                        district_top_ins['District'].append(m['entityName'])
                        district_top_ins['Transaction_type'].append(m['metric']['type'])
                        district_top_ins['Transaction_count'].append(m['metric']['count'])
                        district_top_ins['Transaction_amount'].append(m['metric']['amount'])
                elif x =='pincodes' and y:
                    for l in y:
                        pincode_top_ins['States'].append(i)
                        pincode_top_ins['Years'].append(j)
                        pincode_top_ins['Quarter'].append(int(k.strip('.json')))
                        pincode_top_ins['Pincodes'].append(l['entityName'])
                        pincode_top_ins['Transaction_type'].append(l['metric']['type'])
                        pincode_top_ins['Transaction_count'].append(l['metric']['count'])
                        pincode_top_ins['Transaction_amount'].append(l['metric']['amount'])
#Succesfully created a dataframe
top_insurance_district=pd.DataFrame(district_top_ins)
top_insurance_pincode =pd.DataFrame(pincode_top_ins)
top_insurance_pincode
```
